# Results

```python
import pandas as pd
import numpy as np

def extract_data(file_name, show_to_docfile):
    file = pd.read_csv(file_name)

    sorted_by_first = file.sort_values(by=['Mixed metric roc_auc'], ascending=False)
    sorted_by_first = sorted_by_first[["Dataset", "Model", "Filters", "Baseline value", 'Mixed metric roc_auc']]
    imporovement = []
    for id, row in sorted_by_first.iterrows():
        imporovement.append(round(row['Mixed metric roc_auc']/row['Baseline value'] * 100, 4))

    sorted_by_first['Predicted ROC-AUC'] = round((sorted_by_first['Mixed metric roc_auc'] - sorted_by_first['Baseline value'])*100, 4)
    sorted_by_first['Baseline value']= round(sorted_by_first['Baseline value']*100, 4)
    sorted_by_first = sorted_by_first.drop('Mixed metric roc_auc', axis=1)

    sorted_by_first['Ratio [%]'] = imporovement


    models = np.unique(sorted_by_first['Model'])
    
    sorted_by_first.to_csv("out_"+file_name, sep=';')

    lists_by_models = list(map(lambda x: sorted_by_first[sorted_by_first['Model'] == x], models))

    if show_to_docfile:
        sorted_by_first = sorted_by_first.drop('Dataset', axis=1)
        sorted_by_first = sorted_by_first.drop('Model', axis=1)
    pd.set_option('display.width', 1000)
    print("General: ")
    print(sorted_by_first.iloc[:5])
    for list_by_model, model_name in zip(lists_by_models, models):
        if show_to_docfile:
            list_by_model = list_by_model.drop('Dataset', axis=1)
            list_by_model = list_by_model.drop('Model', axis=1)
        print(model_name)
        print(list_by_model.iloc[:5])
        list_by_model.to_csv("out_different_"+model_name+"_"+file_name, sep=';')
```

## Results for classification BBBP

## Regression

```python
def extract_data_for_regression(file_name, show_to_docfile):
    file = pd.read_csv(file_name)

    sorted_by_first = file.sort_values(by=['Mixed RMSE'], ascending=False)
    sorted_by_first = sorted_by_first[["Dataset", "Model", "Filters", "Baseline RMSE", 'Mixed RMSE']]
    imporovement = []
    for id, row in sorted_by_first.iterrows():
        imporovement.append((row['Mixed RMSE']/row['Baseline RMSE'] * 100))

    sorted_by_first['Predicted RMSE'] = np.round(sorted_by_first['Mixed RMSE'], 4)
    sorted_by_first = sorted_by_first.drop('Mixed RMSE', axis=1)
    sorted_by_first['Baseline RMSE'] = np.round(sorted_by_first['Baseline RMSE'], 4)

    sorted_by_first['Ratio [%]'] = imporovement
    sorted_by_first = sorted_by_first.sort_values(by=['Ratio [%]'], ascending=True)

    models = np.unique(sorted_by_first['Model'])
    
    sorted_by_first.to_csv("out_"+file_name, sep=';')

    lists_by_models = list(map(lambda x: sorted_by_first[sorted_by_first['Model'] == x], models))

    if show_to_docfile:
        sorted_by_first = sorted_by_first.drop('Dataset', axis=1)
        sorted_by_first = sorted_by_first.drop('Model', axis=1)
    print("General:")
    print(sorted_by_first.iloc[:5])
    for list_by_model, model_name in zip(lists_by_models, models):
        if show_to_docfile:
            list_by_model = list_by_model.drop('Dataset', axis=1)
            list_by_model = list_by_model.drop('Model', axis=1)
        print(model_name)
        print(list_by_model.iloc[:5])
        list_by_model.to_csv("out_different_"+model_name+"_"+file_name, sep=';')
```

## Results for regression ESOL

```python
extract_data_for_regression("results_for_regression_with_1_filter.csv", True)
```

```python
extract_data_for_regression("results_for_regression_with_2_filters.csv", True)
```
