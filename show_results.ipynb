## For one filter

```python
import pandas as pd
```

```python
file = pd.read_csv("results_for_classification.csv")
```

```python
for line in file.iterrows():
    print(line[1]['Baseline value'])
    print(line[1]['After filter application'])
    print(line[1]['Mixed metric roc_auc'])
    print(line[1]['Rest metric roc_auc'])
```

```python
sorted_by_first = file.sort_values(by=['After filter application'], ascending=False).iloc[:20]
sorted_by_first = sorted_by_first[["Dataset", "Model", "Filter type", "Baseline value", 'After filter application', 'Mixed metric roc_auc', 'Rest metric roc_auc']]
print(sorted_by_first)
```

```python
sorted_by_second = file.sort_values(by=['Mixed metric roc_auc'], ascending=False).iloc[:20]
sorted_by_second = sorted_by_second[["Dataset", "Model", "Filter type", "Baseline value", 'After filter application', 'Mixed metric roc_auc', 'Rest metric roc_auc']]
print(sorted_by_second)
```

```python
sorted_by_third = file.sort_values(by=['Rest metric roc_auc'], ascending=False).iloc[:20]
sorted_by_third = sorted_by_third[["Dataset", "Model", "Filter type", "Baseline value", 'After filter application', 'Mixed metric roc_auc', 'Rest metric roc_auc']]
print(sorted_by_third)
```
